1-1. 데이터 읽어 DataFrame으로 변경

```python
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from sklearn.preprocessing import MinMaxScaler

# Matplotlib에서 마이너스 기호가 제대로 보이도록 설정
plt.rc('axes', unicode_minus=False)
```

```python
# load dataset/txt를 읽어 탭으로 구분된 데이터 파일을 읽어 pandas DataFrame으로 변환
normal_df = pd.read_table("BC-TCGA-Normal.txt")
tumor_df = pd.read_table("BC-TCGA-Tumor.txt")
```

```python
#tumor DataFrame으로 변환 확인
tumor_df
```

```python
#nomal DataFrame으로 변환 확인
normal_df.head()
```

```python
#tumor,nomal의 행(row)과 열(column)의 개수 확인
print("tumor:",tumor_df.shape)
print("normal:",normal_df.shape)
```

1-2. feature을 열로 바꾸기 위해 전치

```python
#tumor,nomal 전치(transpose)
tumor_df_T = tumor_df.T
normal_df_T = normal_df.T
```

```python
#tumor DataFrame 전치 확인
tumor_df_T.head()
```

```python
#normal DataFrame 전치 확인
normal_df_T
```

```python
tumor_df_T.rename(columns=tumor_df_T.iloc[0], inplace=True)
#데이터프레임의 첫 번째 행 값을 열 이름으로 데이터프레임의 열 이름을 변경하고 변경된 바를 원래 데이터프레임에 적용
tumor_df_T = tumor_df_T.drop(tumor_df_T.index[0])#특징 중복 제거
tumor_df_T.head()
```

```python
normal_df_T.rename(columns=normal_df_T.iloc[0], inplace=True)
#데이터프레임의 첫 번째 행 값을 열 이름으로 데이터프레임의 열 이름을 변경하고 변경된 바를 원래 데이터프레임에 적용
normal_df_T = normal_df_T.drop(normal_df_T.index[0])#특징 중복 제거
normal_df_T.head()
```

```python
#레이블 추가하기: tumor=1, normal=0 즉, 암 발생시 1

tumor_df_T['label'] = 1
normal_df_T['label'] = 0
```

```python
normal_df_T.columns, tumor_df_T.columns #각각 label 추가된 것 확인
```

```python
tumor_df_T.info(), normal_df_T.info()#각각의 정보 확인
```

```python
import sklearn
#두 데이터셋 합치기
df = pd.concat([normal_df_T, tumor_df_T])
df = sklearn.utils.shuffle(df)#무작위로 섞기/ 학습 시 데이터를 더 다양하게 하고 일반화 성능을 향상
df#확인 190개로 합쳐짐
```

```python
df.reset_index(inplace=True)
#데이터프레임의 인덱스가 0부터 시작하는 새로운 정수 인덱스로 재설정
df.drop(['index'],axis=1,inplace=True)
#사람 구분자 삭제
```

```python
df.head()# 확인
```

```python
#df.to_csv("cancer_calssification.csv",index=False)
#제작된 dataframe 저장 이후 프로그램에서는 이를 불러옴
```

```python
df_checknull = pd.DataFrame(df.isnull().sum())#각 열의 null 개수 데이터프레임
df_havenull = df_checknull[df_checknull[0]!=0]# null 값이 있는 열 데이터프레임
havenull_columns = list(df_havenull.index)# null 값이 있는 열의 이름 리스트
df_havenull#확인
```

```python
for col in havenull_columns:#null 값이 있는 열들에 대해서
    df.loc[(df[col].isnull()) & (df['label']==1),col] = df[df['label']==1][col].mean()
    #label이 1이면 label이 1인 행들 중 해당 열의 평균값으로 대체
    df.loc[(df[col].isnull()) & (df['label']==0),col] = df[df['label']==0][col].mean()
    #label이 0이면 label이 0인 행들 중 해당 열의 평균값으로 대체
```

```python
df.isnull().values.any()#null이 모두 제거되었는 지 확인
```

```python
# 정규화
scaler = MinMaxScaler()
scaler.fit(df.iloc[:, :-1])
data_scaled = scaler.transform(df.iloc[:, :-1])
data = pd.DataFrame(data=data_scaled, columns=df.iloc[:, :-1].columns)
data
```

```python
'''
import seaborn as sns
import matplotlib.pyplot as plt
corr_matrix = data.corr()
print(corr_matrix)'''
```

```python
new_df = pd.concat([df.loc[:, 'label'], data], axis=1)#정규화한 feature와 label 합치기
new_df#확인
```

```python
rf = RandomForestClassifier(n_jobs=-1, class_weight='balanced', max_depth=5)
boruta = BorutaPy(rf, n_estimators='auto', verbose=2, random_state=42)#BurutaPy 생성

stats=data.iloc[:, :-1].describe()#정규화 한 값 분석
stats
```

# BurutaPy 실행 및 분석

```python
X = new_df.iloc[:, :-1].values
y = new_df.loc[:, 'label']
#X, y값 정하기
# borutaPy 동작
boruta.fit(X, y)
#selected_features = df.columns[:-1][boruta.support_]
```

```python
# 결과 출력
print(boruta.support_)#True=유의미/False=무의미
print(boruta.ranking_)#작을수록 의미
print(boruta.n_features_)#유의미한 feature 개수
selected_features = data.columns[boruta.support_]#유의미한 feature 이름
print(selected_features[:])#확인
```
