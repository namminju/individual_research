```python
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from sklearn.preprocessing import MinMaxScaler

# Matplotlib에서 마이너스 기호가 제대로 보이도록 설정
plt.rc('axes', unicode_minus=False)
```

```python
# load dataset/txt를 읽어 탭으로 구분된 데이터 파일을 읽어 pandas DataFrame으로 변환
normal_df = pd.read_table("BC-TCGA-Normal.txt")
tumor_df = pd.read_table("GBM.txt")
```

1-2. feature을 열로 바꾸기 위해 전치

```python
#tumor,nomal 전치(transpose)
tumor_df_T = tumor_df.T
normal_df_T = normal_df.T
```

```python
#tumor DataFrame 전치 확인
tumor_df_T.head()
```

```python
#normal DataFrame 전치 확인
normal_df_T
```

```python
tumor_df_T.rename(columns=tumor_df_T.iloc[0], inplace=True)
#데이터프레임의 첫 번째 행 값을 열 이름으로 데이터프레임의 열 이름을 변경하고 변경된 바를 원래 데이터프레임에 적용
tumor_df_T = tumor_df_T.drop(tumor_df_T.index[0])#특징 중복 제거
tumor_df_T.head()
```

```python
normal_df_T.rename(columns=normal_df_T.iloc[0], inplace=True)
#데이터프레임의 첫 번째 행 값을 열 이름으로 데이터프레임의 열 이름을 변경하고 변경된 바를 원래 데이터프레임에 적용
normal_df_T = normal_df_T.drop(normal_df_T.index[0])#특징 중복 제거
normal_df_T.head()
```

```python
#레이블 추가하기: tumor=1, normal=0 즉, 암 발생시 1

tumor_df_T['label'] = 1
normal_df_T['label'] = 0
```

```python
common_cols = list(set(normal_df_T.columns) & set(tumor_df_T.columns))
normal_df_T = normal_df_T[common_cols]
tumor_df_T = tumor_df_T[common_cols]
#feature 겹치는 것만 추출
```

```python
normal_df_T.columns, tumor_df_T.columns #추출 잘 됐는 지 각각 label 추가된 것 확인
```

```python
tumor_df_T.info(), normal_df_T.info()#각각의 정보 확인
```

```python
#tumor,nomal의 행(row)과 열(column)의 개수 확인
print("tumor:",tumor_df_T.shape)
print("normal:",normal_df_T.shape)
```

```python
normal_df_T.head()
```

```python
tumor_df_T.head()
```

```python
import sklearn
#두 데이터셋 합치기
df = pd.concat([normal_df_T, tumor_df_T])
df = sklearn.utils.shuffle(df)#무작위로 섞기/ 학습 시 데이터를 더 다양하게 하고 일반화 성능을 향상
df#확인 502개로 합쳐짐
```

```python
df.reset_index(inplace=True)
#데이터프레임의 인덱스가 0부터 시작하는 새로운 정수 인덱스로 재설정
df.drop(['index'],axis=1,inplace=True)
#사람 구분자 삭제
```

```python
df.head()# 확인
```

```python
label_rows = df['label']
label_rows
#사이에 숨어버린 label 추출
```

```python
# 'label' 열 제거
new_df = df.drop('label', axis=1)
```

```python
#다시 함쳐 label의 위치가 맨 뒤로 가게 변경
df = pd.concat([new_df, label_rows], axis=1)
df.head()
```

```python
df.to_csv("cancer.csv",index=False)
#제작된 dataframe 저장 이후 프로그램에서는 이를 불러옴
```
